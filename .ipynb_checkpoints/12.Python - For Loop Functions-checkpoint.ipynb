```python
def find_len(s):
    if s == "":
        return 0
    else :
        return 1+ find_len(s[1:])
```

```python
find_len("sudhanshu")
```

```python
find_len("pwskills")
```

```python
def find_reverse(s):
    if s=="":
        return ""
    else:
        return find_reverse(s[1:]) + s[0]
```

```python
find_reverse("sudh")
```

```python
find_reverse("pwskills")
```

```python
s = "sfafafdaaf"
```

```python
def find_char_count(s,c):
    if s=="":
        return 0
    elif s[0] == c:
        return 1+find_char_count(s[1:], c)
    else :
        return find_char_count(s[1:],c)
```

```python
find_char_count(s,'f')
```

```python
find_char_count(s,'a')
```

```python
find_char_count(s,'d')
```

```python
l = [4,5,6,3,4,5,6,6,7]
```

```python
def check_sort_list(l):
    if len(l) <=1:
        return True
    elif l[0] > l[1]:
        return False
    else :
        return check_sort_list(l[1:])
```

```python
check_sort_list(l)
```

```python
def check_per_square(n,start =1):
    if n== start **2:
        return True
    elif n< start **2:
        return False
    else:
        return check_per_square(n,start+1)
```

```python
check_per_square(64)
```

```python
check_per_square(6)
```
