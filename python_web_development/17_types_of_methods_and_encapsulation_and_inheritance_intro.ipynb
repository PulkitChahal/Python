```python
class student:
    
    school = 'ineuron' # class variable
    
    def __init__(self):
        self.marks = 10 # instance variable
        
    def get_marks(self): # instance methods
        return self.marks
    
    @classmethod
    def get_school(cls): # class methods
        return cls.school
    
    @staticmethod
    def add(x, y): # static methods
        return x+y

s1 = student()

print(s1.marks)

print(student.school)

print(s1.get_school())
print(student.get_school())

print(student.add(3,4))
print(s1.add(3,4))
```

```python
# Instance method

class MyClass:
    def __init__(self, value):
        self.value = value
    
    def instance_method(self):
        return f'Instance method called, value is {self.value}'

obj = MyClass(10)
print(obj.instance_method())  # Output: Instance method called, value is 10
```

```python
# Class method

class MyClass:
    class_attr = 0
    
    @classmethod
    def class_method(cls):
        cls.class_attr += 1
        return f'Class method called, class_attr is {cls.class_attr}'

print(MyClass.class_method())  # Output: Class method called, class_attr is 1
```

```python
# Static Method

class MyClass:
    
    @staticmethod
    def static_method(arg):
        return f'Static method called with argument {arg}'

print(MyClass.static_method(42))  # Output: Static method called with argument 42
```

```python
# Special method

class MyClass:
    def __init__(self, value):
        self.value = value
    
    def __str__(self):
        return f'MyClass instance with value {self.value}'
    
    def __add__(self, other):
        return MyClass(self.value + other.value)

obj1 = MyClass(10)
obj2 = MyClass(20)
obj3 = obj1 + obj2
print(obj3)  # Output: MyClass instance with value 30
```

```python
class person:
    
    def __init__(self):
        self.age = 10
        
    def get_age(self):
        return self.age
        
    def set_age(self, age):
        self.age = age
        
        
p1 = person()

print(p1.age)

print(p1.get_age())

p1.set_age(19)
print(p1.get_age())
```

```python
# Property method

class MyClass:
    def __init__(self, value):
        self._value = value
    
    @property
    def value(self):
        return self._value
    
    @value.setter
    def value(self, new_value):
        if new_value >= 0:
            self._value = new_value
        else:
            raise ValueError("Value must be non-negative")

obj = MyClass(10)
print(obj.value)  # Output: 10
obj.value = 20
print(obj.value)  # Output: 20
# obj.value = -10  # This will raise ValueError
```

```python
# __age -> data hiding
# _age ->
class person:
    
    def __init__(self):
        self.__age = None
        self._name = 'pulkit'
        
    def get_age(self):
        return self.__age
        
    def set_age(self, age):
        self.__age = age
        
         
p1 = person()
p1.set_age(19)
print(p1.get_age())

print(p1._name)

p1.__age = 20
print(p1.__age)

print(p1.get_age())
```

```python
# __age -> data hiding
# _age ->
class person:
    
    def __init__(self):
        self.__age = None
        self._name = 'pulkit'
        
    def __private_method(self):
        return f'Value is {self.__age}'
    
    def get_age(self):
        return self.__age
        
    def set_age(self, age):
        self.__age = age
        
        
p1 = person()
p1.set_age(19)

print(p1._person__age)
print(p1._person__private_method())
```

```python
# Public

class MyClass:
    def __init__(self, value):
        self.value = value
    
    def public_method(self):
        return f'Value is {self.value}'

obj = MyClass(10)
print(obj.value)           # Output: 10
print(obj.public_method()) # Output: Value is 10
```

```python
# Private

class MyClass:
    def __init__(self, value):
        self.__value = value
    
    def __private_method(self):
        return f'Value is {self.__value}'
    
    def public_method(self):
        return self.__private_method()

obj = MyClass(10)
# print(obj.__value)              # AttributeError
# print(obj.__private_method())   # AttributeError
print(obj.public_method())        # Output: Value is 10

# Accessing private members through name mangling
print(obj._MyClass__value)         # Output: 10
print(obj._MyClass__private_method())  # Output: Value is 10
```

```python
# Protected

class MyClass:
    def __init__(self, value):
        self._value = value
    
    def _protected_method(self):
        return f'Value is {self._value}'
    
    def public_method(self):
        return self._protected_method()

class SubClass(MyClass):
    def access_protected(self):
        return self._protected_method()

obj = MyClass(10)
print(obj._value)                # Output: 10
print(obj._protected_method())   # Output: Value is 10

sub_obj = SubClass(20)
print(sub_obj.access_protected()) # Output: Value is 20
```

```python
class iphone13:
    
    def f1(self):
        print('in iphone13 f1')
        
        
class iphone14(iphone13):
    
    def f2(self):
        print('in iphone14 f2')

        
class iphone15(iphone14):
    
    def f3(self):
        print('in iphone15 f3')

i_13 = iphone13()
i_13.f1()
# i_13.f2() # error
# i_13.f3() #error
print('----------')

i_14 = iphone14()
i_14.f1()
i_14.f2()
# i_14.f3() # error
print('----------')

i_15 = iphone15()
i_15.f1()
i_15.f2()
i_15.f3()
```
