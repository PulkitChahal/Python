```python
class os1:
    def call(self):
        print('calling')
        
    def get_notification(self):
        print('notification on top, 15')
        
class os2(os1):
    def get_notification(self):
        print('notification at bottom, 16')
        
    def dance(self):
        print('dancing')

class iphone:
    def get_osversion(self, os):
        os.get_notification()
        
    def call(self, os):
        os.call()
        
os = os1()
os = os2()

obj = iphone()
obj.get_osversion(os)
obj.call(os)
```

```python
class calculator:
    def add(self, n1, n2=None, n3=2):
        s = 0
        if n1 != None and n2 != None and n3 != None:
            s = n1 + n2 + n3
        elif n1 != None and n2 != None:
            s = n1 + n2
        elif n1 != None and n3 != None:
            s = n1 + n3
        elif n2 != None and n3 != None:
            s = n2 + n3
        elif n3 != None:
            s = n1 + n3
        elif n2 != None:
            s = n1 + n2         
        elif n1 != None:
            s = n2 + n3
        else:
            s = n1
        return s
        
c1= calculator()
c1.add(4)
```
