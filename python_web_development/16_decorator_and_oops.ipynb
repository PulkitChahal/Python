```python
def decor_result(result_func):
    def distinction(marks):
        for m in marks:
            if m >= 75:
                print('Distinction')
        else:
            result_func(marks)
    return distinction

@decor_result
def result(marks):
    for m in marks:
        if m > 33:
            pass
        else:
            print('Fail')
            break
    else:
        print('Pass')
        
result([55,60,70,75,80])
```

```python
class computer:
    
    def __init__(self, p, r):
        self.cpu = p
        self.ram = r
    
    def show(self):
        print(self.cpu, self.ram)

        
com1 = computer('i7', 16)
com2 = computer('i9', 32)

com1.show() # computer.show(com1)
com2.show() # computer.show(com2)
```

```python
class computer:
    
    brand = 'HP'
    
    def __init__(self, p, r):
        self.cpu = p
        self.ram = r
        self.brand = self.brand
    
    def show(self):
        print(f'cpu : {self.cpu}, ram : {self.ram}, brand : {self.brand}')

        
com1 = computer('i7', 16)
com2 = computer('i9', 32)

print(computer.brand)

com1.show() # computer.show(com1)
com2.show() # computer.show(com2)
```

```python
class student:
    
    def __init__(self,marks):
        self.marks=marks
        
    def compare(s1, s2):
        if s1.marks > s2.marks:
            print('s1')
        else:
            print('s2')
        
        
s1=student(25)
s2=student(30)

student.compare(s1,s2)

s1.compare(s2)
s2.compare(s1)
```
