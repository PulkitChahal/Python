```python
import threading

def print_numbers():
    for i in range(10):
        print(i)

# Create a thread
thread = threading.Thread(target=print_numbers)
# thread = Thread(target=print_numbers)

# Start the thread
thread.start()

# Wait for the thread to complete
thread.join()
```

```python
import threading

lock = threading.Lock()

def print_numbers():
    for i in range(10):
        lock.acquire()
        try:
            print(i)
        finally:
            lock.release()

thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_numbers)

thread1.start()
thread2.start()

thread1.join()
thread2.join()
```
