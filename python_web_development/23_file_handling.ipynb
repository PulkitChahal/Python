```python
def writing(filename, text):
    with open(filename, 'w') as f:
        f.write(text)

writing('test.txt', 'hello pulkit')
```

```python
def append(filename, text):
    with open(filename, 'a') as f:
        f.write(text)
        
append('test.txt', '\nwelcome to append function')
```

```python
def reading(filename):
    try:
        with open(filename, 'r') as f:
            text = f.read()
            print(text)
            f.close()
            
    except FileNotFoundError:
        print('file not found')
        
reading('test.txt')
```

```python
def search(filename, word):
    try:
        with open(filename, 'r') as f:
            line_count = 0
            for line in f.readlines():
                line_count += 1
                str_list = line.split(' ')
                word_count = 0
                for w in str_list:
                    word_count += 1
                    if word == w:
                        return (line_count, word_count)
            else:
                return None
        f.close()
            
    except FileNotFoundError:
        print('file not found')
    
search('test.txt', 'append')
```

```python
def modify(filename, old_word, new_word):
    t = search(filename, old_word)
    if t != None:
        mylist = []
        
        try:
            f = open(filename, 'r')
            for line in f.readlines():
                line = line.replace(old_word, new_word)
                mylist.append(line)
            f.close()
            
            f = open(filename, 'w')
            f.write(' '.join(mylist))
            f.close()
            
        except FileNotFoundError:
            print('file not found')
    else:
        print('search failed')
        
modify('test.txt', 'append', 'add')
```
