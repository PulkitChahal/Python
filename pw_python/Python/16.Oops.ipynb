```python
import time
class timer:
    def __init__(self):
        self.start_time = time.time()
        
    def task(self):
        time_spent = time.time() - self.start_time
        print(time_spent)
    
    def __del__(self):
        print("")
    
    def __str__(self):
        return "this is my class timer"
```

```python
import time
def print_list(l):
    start_tiem = time.time()
    for i in l:
        print(i)
    end_time = time.time()
    total_time = end_time - start_time
    print(total_time)
```

```python
def print_key(d):
    start_tiem = time.time()
    print(d.keys())
    end_time = time.time()
    total_time = end_time - start_time
    print(total_time)
```

```python
def find_time(func):
    def cal_time(*args):
        start_time = time.time()
        func(*args)
        end_time = time.time()
        total_time = end_time - start_time
        print(total_time)
        
    return cal_time
```

```python
@find_time
def print_key(d):
    print(d.keys())
```

```python
print_key({"key":"value","name":"sudh"})
```

```python
@find_time
def print_list(l):
    for i in l:
        print(i)
```

```python
print_list([23,4,5,5,6,67,7,8])
```

```python
import logging
def log_func(func):
    def log_inner(*args):
        logging.basicConfig(filename = "test.log", level=logging.INFO)
        logging.info("this is the start of my func")
        func(*args)
        logging.info("this is the end of my func")
    return log_inner
```

```python
@log_func
def print_list(l):
    for i in l:
        print(i)
```

```python
print_list([23,4,5,5,6,67,7,8])
```

```python
@find_time
@log_func
def print_list(l):
    for i in l:
        print(i)
```

```python
print_list([23,4,5,5,6,67,7,8])
```

```python
class sudh:
    def __init__(self, subject):
        self.__subject = subject
        
    @property
    def subject(self):
        return self.__subject
    
    @subject.setter
    def subject(self, subject):
        self.__subject = subject
    
    @subject.getter
    def subject(self):
        return self.__subject
```

```python
s2 = sudh("data analytics")
```

```python
s2.subject
```
