```python
#06 Count the number of consonants in a string.
```

```python
def constant_char(string):
    constant_value = 0
    constant = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"
    for char in string:
        if char in constant:
            constant_value += 1
    return constant_value

input_string = input('Enter string: ')
result = constant_char(input_string)
print(result)
```

```python
#07 Remove all whitespaces from a string.
```

```python
def remove_space(string):
    result = ''
    for char in string:
        if char != " ":
            result += char
    return result

input_string = input("enter string: ")
remove_space(input_string)
```

```python
#08 Find the length of a string without using the `len()` function.
```

```python
def length(string):
    len = 0
    for char in string:
        if char in string:
            len += 1
    return len

input_string = input("enter string: ")
length(input_string)
```

```python
#09 Check if a string contains a specific word.
```

```python
def specific_char(string, word):
    for char in string:
        if word in string:
            return word,"present in string"
        else:
            return word,"not present in string"

input_string = input('Enter String : ')
alphabet = input('Enter Alphabet : ')
specific_char(input_string, alphabet)
```

```python
#10 Replace a word in a string with another word.
```

```python
def replace_word(string, letter ,replace_letter):
    empty_string = ''
    for char in string:
        if char == letter:
            empty_string += replace_letter
        else:
            empty_string += char
    return empty_string

input_string = input('Enter String : ')
word = input('Enter word : ')
replace_char = input('Enter word to replace : ')
replace_word(input_string, word, replace_char)            
```

```python
#11 Count the occurrences of a word in a string.
```

```python
def occur(string, word):
    count = 0
    for char in string:
        if char == word:
            count += 1
    return count

string = input('Enter String : ')
char = input('Enter letter : ')
occur(string, char)
```

```python
#12. Find the first occurrence of a word in a string.
```

```python
def first_occur(string, word):
    index = string.find(word)
    
    if index == -1:
        return None
    
    return index

string = input('Enter String : ')
char = input('Enter letter : ')
first_occur(string, char)            
```

```python
#17 Check if a string starts with a specific word or phrase.
```

```python
def starts(text , prefix):
    words = text.startswith(prefix)
    return words

starts('apple','b')
```

```python
#18 Check if a string ends with a specific word or phrase.
```

```python
def ends(text, word):
    str = text.endswith(word)
    return str

ends('hello boy','b')
```
