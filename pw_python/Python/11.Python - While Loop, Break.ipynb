```python
def fib(n):
    if n<=1:
        return n
    else:
        return fib(n-2) +fib(n-1)
```

```python
def sum_of_digit(n):
    sum =0
    for digit in str(n):
        sum += int(digit)
    return sum
```

```python
sum_of_digit(234)
```

```python
def get_sum_of_digit(n):
    sum=0
    while n!=0:
        d=n%10
        sum=sum+d
        n=n//10
    return sum
```

```python
get_sum_of_digit(234)
```

```python
def sum_digit(n):
    if n <= 9:
        return n
    else :
        return n%10 + sum_digit(n//10)
```

```python
sum_digit(234)
```

```python
def max_find(l):
    if len(l) == 1:
        return l[0]
    else:
        return max(l[0],max_find(l[1:]))
```

```python
max_find([4,5,6,6,7,8,43,1,3,6575,8])
```

```python
def power_fun(n, k):
    if k == 0:
        return 1
    else:
        return n * power_fun(n, k-1)
```

```python
power_fun(2,4)
```

```python
power_fun(3,2)
```
