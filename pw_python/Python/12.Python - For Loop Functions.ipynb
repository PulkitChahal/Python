```python
def find_char_count(s,c):
    if s=="":
        return 0
    elif s[0] == c:
        return 1+find_char_count(s[1:], c)
    else :
        return find_char_count(s[1:],c)
```

```python
find_char_count(s,'f')
```

```python
find_char_count(s,'a')
```

```python
def check_per_square(n,start =1):
    if n== start **2:
        return True
    elif n< start **2:
        return False
    else:
        return check_per_square(n,start+1)
```

```python
check_per_square(64)
```

```python
check_per_square(6)
```

```python
def count_fun(n):
    i = 1
    while i<=n:
        yield i
        i = i+1
```

```python
for i in count_fun(20):
    print(i)
```

```python
def fib(n):
    a,b = 0,1
    for i in range(n):
        yield a
        a,b = b, a+b
```

```python
for i in fib(9):
    print(i)
```

```python
def read_file(file_name):
    with open(file_name) as file :
        for i in file:
            yield i
```

```python
read_file('test.txt')
```

```python
for i in read_file("test.txt"):
    print(i)
```
