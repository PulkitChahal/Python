```python
# Bubble Sort

def bubbleSort(arr):
    for i in range(len(arr)-1, 0 ,-1):
        for j in range(i):
            if (arr[j]>arr[j+1]):
                arr[j],arr[j+1] = arr[j+1],arr[j]
```

```python
arr = [2,5,4,9,3]
bubbleSort(arr)
print(arr)
```

```python
# Bubble Sort

def bubbleSortOptimized(arr):
    for i in range(len(arr)-1, 0 ,-1):
        isSorted = True
        for j in range(i):
            if (arr[j]>arr[j+1]):
                isSorted = False
                arr[j],arr[j+1] = arr[j+1],arr[j]
                
        if isSorted:
            print('Array is already sorted')
            break
```

```python
arr = [2,56,4,9,3]
bubbleSortOptimized(arr)
print(arr)
```

```python
# Given an array, check the array is already sorted

def isSorted(arr):
    isSorted = True
    for i in range(len(arr)-1):
        if arr[i]>arr[i+1]:
            isSorted = False
            break
        if isSorted:
            print('Array is sorted')
        else:
            print('Array is not sorted')
```

```python
# Selection Sort

def slectionsort(arr):
    for i in range(len(arr)):
        min = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[min]:
                min = j
        # Find the samllest values
        arr[i],arr[min] = arr[min],arr[i]
```

```python
# Insertion Sort

def insertionsort(arr):
    for i in range(1,len(arr)):
        v = arr[i]
        j = i
        while (j >= 1 and arr[j-1] > v):
            arr[j]=arr[j-1]
            j-=1
        arr[j]=v
```

```python
# Linear Search

def search(arr, num):
    for i in range(len(arr)):
        if num == arr[i]:
            return i
    return -1
```

```python
print(search([4,5,6,7,8,2,9,3],7))
```

```python
# Binary Search

def binarysearch(arr, num):
    left = 0
    right = len(arr) - 1        
    while(left <= right):
        #mid = (left + right)/2
        mid = int(left + (right - left)/2)
        if arr[mid] == num:
            return mid
        elif arr[mid] > num:
            right = mid-1
        else:
            left = mid+1
    return -1
```
