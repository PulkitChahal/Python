```python
def printRec(arr, sI):
    #Base condition
    if sI >= len(arr):
        return
    
    #Logic
    print(arr[sI])
    
    #Recursive call
    printRec(arr, sI+1)

printRec([1,2,3,4],0)
```

```python
def printRec(arr, sI):
    #Base condition
    if sI >= len(arr):
        return
    
    #Recursive call
    printRec(arr, sI+1)
    
    #Logic
    print(arr[sI])

printRec([1,2,3,4],0)
```

```python
def findmin(arr, s):
    #Base condition
    if s >= len(arr):
        return 999999
    
    return min(arr[s], findmin(arr,s+1))

print(findmin([3,5,2],0))
```

```python
def moves(n, s, d, a):
    # Base condition
    if n == 1:
        return 1
    # move n-1 s->a, then move 1 biggest coin to s->d and then n-1 coin a->d
    return moves(n-1,s,a,d) + 1 + moves(n-1,a,d,s)

print(moves(1,'s','d','a'))
print(moves(5,'s','d','a'))
print(moves(10,'s','d','a'))
```
