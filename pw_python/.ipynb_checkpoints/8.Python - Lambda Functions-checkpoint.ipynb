```python
# Positional Arguments

def add_number(a,b):
    sum =a+b
    print("sum",sum)
```

```python
add_number(2,5)
```

```python
def emp_name(name):
    print("my name is",name)
```

```python
emp_name("imran")
```

```python
# Default Arguments

def add_number(a=4,b=8):
    sum = a+b
    print("sum",sum)
```

```python
add_number()
```

```python
add_number(2,8)
```

```python
def emp_name(name="pwskills"):
    print("my name is",name)
```

```python
emp_name()
```

```python
emp_name("imran")
```

```python
# Keyword Arguments

def display_info(first_name,last_name):
    print("first name",first_name)
    print("last name",last_name)
```

```python
display_info(first_name='pw',last_name='skills')
```

```python
display_info(last_name='skills',first_name='pw')
```

```python
def add_numbers(a,b):
    sum = a+b
    print("sum",sum)
```

```python
add_numbers(a=4,b=7)
```

```python
add_numbers(b=7,a=4)
```

```python
def add_numbers(a,b,c,d):
    sum = a+b+c+d
    print("sum",sum)
```

```python
add_numbers(1,2,3,4)
```

```python
# *args and **kargs
```

```python
def find_sum(*numbers):
    result=0
    
    for num in numbers:
        result =result+num
    print("sum",result)
```

```python
find_sum()
```

```python
find_sum(1,2,3)
```

```python
find_sum(1,2,3,4,5,6,7,8)
```

```python
def intro(**data):
    print("data type of argument",type(data))
    
    for key,value in data.items():
        print(key,value)
```

```python
intro(firstname='pw',lastname='skills')
```
