```python
class pwskills:
    def __init__(self, mentor):
        self.mentor = mentor
    
    def mentor_name(self):
        print(self.mentor)

class datascience(pwskills):
    def __init__(self, mentor, mentor_mail_id):
        self.mentor = mentor
        self.mentor_mail_id = mentor_mail_id
    
    def show_info(self):
        print(self.mentor, self.mentor_mail_id)
```

```python
python_basic = datascience("sudh", "sdfg@gmail.com")
```

```python
python_basic.show_info()
```

```python
python_basic.mentor_name()
```

```python
class human:
    def __init__(self):
        pass
    
    def eat(self):
        print("print the eat method from human")
```

```python
class male(human):
    def __init__(self, name):
        self.name = name
    
    def eat(self):
        super().eat()
        print(self.name)
```

```python
sudh = male("sudhanshu")
```

```python
sudh.eat()
```

```python
class fileopener:
    def __init__(self, filename):
        self.filename = filename
        
    def open_file(self):
        print("this will open the file ", self.filename)
        
    def __del__(self):
        self.filename
```

```python
f2 = fileopener("f2text")
```

```python
f2.open_file()
```

```python
import time
class timer:
    def __init__(self):
        self.start_time = time.time()
        
    def task(self):
        time_spent = time.time() - self.start_time
        print(time_spent)
    
    def __del__(self):
        print("")
    
    def __str__(self):
        return "this is my class timer"
```

```python
t1  = timer()
```

```python
t1.task()
```

```python
print(t1)
```

```python
def test(func):
    def inner_test():
        print("this is the start of my inner test")
        func()
        print("this is the end of my inner test")
    
    return inner_test

@test
def test1():
    print("this is my test1")
```
