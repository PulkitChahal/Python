```python
# merge and join operation
```

```python
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]
                   })
```

```python
df1
```

```python
df2 = pd.DataFrame({'key1':[1,2,45,6,67],
                   'key4':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]
                   })
```

```python
df2
```

```python
# merge operation

pd.merge(df1,df2, how = 'left')
```

```python
pd.merge(df1,df2, how = 'right')
```

```python
pd.merge(df1,df2, how = 'outer') # all the values of both the dataframe
```
