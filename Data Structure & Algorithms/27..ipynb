```python
def search_rec(arr, si, data):
    # Base condition
    if si >=len(arr):
        return -1
    if arr[si] == data:
        return si
    else:
        return search_rec(arr,si+1,data)
    
print(search_rec([4,5,3,1,2],0,7))
```

```python
def binary_search(arr, si, ei, data):
    # Base condition
    if si > ei:
        return -1

    mid = int(si+(ei-si)/2)
    if arr[mid] == data:
        return mid
    elif arr[mid] > data:
        return binary_search(arr, si, mid-1, data)
    else:
        return binary_search(arr, mid+1, ei, data)
    
print(binary_search([2,4,5,8,10],0,9,8))
```

```python
def reverseArr(arr):
    i=0
    j=len(arr)-1
    
    while(i<j):
        #swap elements ar i and j index
        arr[i],arr[j] = arr[j],arr[i]
        i +=1
        j -=1
        
arr = [4,5,1,2,3]
reverseArr(arr)
print(arr)
```

```python
def reverse_rec(arr,si,ei):
    if si>=ei:
        return
    arr[si],arr[ei]=arr[ei],arr[si]
    reverse_rec(arr,si+1,ei-1)
    
arr = [1,2,3,4,5]
reverse_rec(arr, 0, len(arr)-1)
print(arr)
```

```python
def palindrome(arr):
    # 2 pointer approach
    left = 0
    right =len(arr)-1
    
    while(left< right):
        if arr[left] != arr[right]:
            return False
     
        left +=1
        right -=1
    return True

arr = [1,2,3,3,2,1]
arr1=[1,2,4]
arr3=[1]

print(palindrome(arr))
print(palindrome(arr1))
print(palindrome(arr3))
```

```python
def rotate(arr, k):
    k = k%len(arr)
    # num of times the rotation has to take place
    for i in range(k):
        # Logic to rotate the array
        prev = arr[0]
        curr = arr[0]
        arr[0] = arr[len(arr)-1]
        j=1
        while(j<len(arr)):
            curr =arr[j]
            arr[j] = prev
            prev =curr
            j+=1
            
arr1=[1,2,3,4,5]
rotate(arr1,1)
print(arr1)
```

```python
def rotate_k_times(arr, k):
    k= k% len(arr)
    
    # reverse the entire array
    reverse_rec(arr,0,len(arr)-1)
    
    # reverse the first k elements
    reverse_rec(arr,0,k-1)
    
    # rverese the last n-k elements
    reverse_rec(arr, k, len(arr)-1)
```
