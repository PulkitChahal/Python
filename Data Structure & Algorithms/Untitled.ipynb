```python
## Factorial of a given number
## Method definition via the recursion

def factorial(n):
    ## 1. Base case condition
    if n <= 1:
        return 1
    
    ## 2. Recursive function call
    return n * factorial(n-1)

## Driver code
n = 5
result = factorial(n)
print(result)
```

```python
## Method definition
## Time complexity: O(n)
## Space complexity: O(n) [n represents numbe rof digits]

def sum_of_digits(n):
    ## base case condition
    if n < 10:
        return n
    
    ## recursive function call
    return n % 10 + sum_of_digits(n // 10)
    
## Driver code
n = 1234
result = sum_of_digits(n)
print(result)
```

```python
## Method definition
## Time complexity: O(n)
## Space complexity: O(n)

def powerOf(a,n):
    ## base case condition
    if n == 0:
        return 1
    else:
    ## recursive function call 
        return a * powerOf(a,(n-1))
    
## Driver code
a = 2 
n = 16
result = powerOf(a,n)
print(result)
```

```python
## Method definition

def powerFind(a, n):
    ## base case conditiom
    if n == 1:
        return a
    
    ## recurcive call function
    else:
        mid = n // 2
        result = powerFind(a, mid)
        finalResult = result * result
        
        ##checking whether the value of n is even or odd
        if n % 2 == 0:
            return finalResult
        else:
            return a * finalResult

## Driver code
a = 2
n = 17
result  = powerFind(a, n)
print(result)
```
