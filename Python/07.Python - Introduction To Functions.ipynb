```python
def cal(a,b):
    oper = input("enetr operation: ")
    if (oper == '+'):
        return a+b
    elif (oper == '-'):
        return a-b
    elif (oper == '*'):
        return a*b
    elif (oper == '/'):
        return a/b
    elif (oper == '%'):
        return a%b
    else:
        return "error"
```

```python
cal(4,5)
```

```python
# Create a python program that includes a user-defined function called print_even_number().
# the function should prompt the user to set a range limit befor printing
# all even numbers upto that list
```

```python
def print_even_number():
    a=int(input("limit first: "))
    b=int(input("limit second: "))
    for i in range(a,b+1):
        if i%2 == 0:
            print(i)
print_even_number()
```

```python
#
def pythagorus():
    p= int(input("side p: "))
    b= int(input("side b: "))
    h= int(input("side h: "))
    if (p**2 == b**2 +h**2) or (b**2 == h**2 + p**2) or (h**2 == p**2 + b**2):
        print("right angled triangle")
    else:
        print("not right angled triangle")
pythagorus()
```
